```python
def img():
  img_RGB2 = cv2.imread("drive/MyDrive/opencv/img/FP0.JPG",1)
  y, x = img_RGB2.shape[:2]
  centro = (int(x/2), int(y/2))

  #cv2.line(img_RGB2, (0, 0), (x, y), (255, 255, 255), 2)
  #cv2.line(img_RGB2, (x, 0), (0, y), (255,255,255), 2)
  #cv2.line(img_RGB2, centro, (centro[0], 0), (255, 255, 255), 2)
  #cv2.line(img_RGB2, centro, (x, centro[1]), (255, 255, 255), 2)
  #cv2.line(img_RGB2, (0,0), (0, y),(0,0,255), 2)
  #cv2.line(img_RGB2, (0,0), (x, 0),(255,0,0), 2)
  #cv2.circle(img_RGB2, centro, (centro[1]), (130, 130, 130), 2)
  #cv2.line(img_RGB2, (0,y), (x, y),(255,0,0), 2)
  #cv2.line(img_RGB2, (x,0), (x, y),(0,0,255), 2)
  #cv2_imshow(img_RGB2)

  return (img_RGB2, x, y)


def trazar_triangulo(img_o, x_l, y_l, x_o, y_o):
    #Triangulo Superior Izquierdo
    #cv2.line(img_o, (0,0), (x_l, 0), (132,129,128), 2) #Base
    #cv2.line(img_o, (0,0), (0, y_l), (132,129,128,2)) #Altura
    #cv2.line(img_o, (0, y_l), (x_l, 0), (0,0,255), 2) #Hipotenusa

    pts = np.array([[0,0],[x_l, 0],[0,y_l]], np.int32)
    cv2.polylines(img_o,[pts],True,(0,255,255))


    #Triangulo Superior Derecho
    #cv2.line(img_o, (x_l,0), (x_o, 0), (255, 100, 0), 2) #Base
    #cv2.line(img_o, (x_o,0), (x_o, y_o-y_l), (255, 100, 0), 2) #Altura
    #cv2.line(img_o, (x_l,0), (x_o, y_o-y_l), (0,0,255), 2) #Hipotenusa

    pts = np.array([[x_o, 0], [x_o,y_o-y_l], [x_l,0]], np.int32)
    cv2.polylines(img_o,[pts],True,(0,0,255))

    #Triangulo Inferior Izquierdo
    #cv2.line(img_o, (0, y_o), (x_o - x_l, y_o), (255, 100, 0), 2) #Base
    #cv2.line(img_o, (0,y_o), (0, y_l), (255, 100, 0), 2) #Altura
    #cv2.line(img_o, (0, y_l), (x_o-x_l, y_o), (0,0,255), 2) #Hipotenusa


    pts = np.array([[0,y_l],[0, y_o],[x_o-x_l,y_o]], np.int32)
    cv2.polylines(img_o,[pts],True,(255,0,0))

    #Triangulo Inferior Derecho
    #cv2.line(img_o, (x_o -x_l, y_o), (x_o, y_o), (132,129,128), 2) #Base
    #cv2.line(img_o, (x_o, y_o-y_l), (x_o,  y_o), (132,129,128), 2) #Altura
    #cv2.line(img_o, (x_o, y_o-y_l), (x_o-x_l, y_o), (0,0,255),2) #Hipotenusa
    pts = np.array([[x_o,y_o-y_l],[x_o, y_o],[x_o-x_l,y_o]], np.int32)
    cv2.polylines(img_o,[pts],True,(0,255,0))
    return img_o

  
```
